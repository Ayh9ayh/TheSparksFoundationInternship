In this task involving K-means clustering, my aim was to determine the ideal number of clusters and visualize them using the 'Iris' dataset.

The goal was to analyze the dataset and find the most suitable number of clusters that best represent the underlying patterns in the data. Additionally, I created visual representations to illustrate the clustering results effectively.

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.patches as mpatches
import sklearn.metrics as sm
from mpl_toolkits.mplot3d import Axes3D
from scipy.cluster.hierarchy import linkage,dendrogram
from sklearn.cluster import DBSCAN 
from sklearn.decomposition import PCA 
```

```python
iris = datasets.load_iris()
print(iris.data)
```

```python
print(iris.target_names)
```

```python
print(iris.target)
```

```python
x = iris.data
y = iris.target
```

```python
#Plotting
fig = plt.figure(1, figsize=(7,5))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(x[:, 3], x[:, 0], x[:, 2], edgecolor="k", s=50)
ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
plt.title("Iris Clustering K Means=3", fontsize=14)
plt.show()

#Hierachy Clustering 
hier=linkage(x,"ward")
max_d=7.08
plt.figure(figsize=(15,8))
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Species')
plt.ylabel('distance')
dendrogram(
    hier,
    truncate_mode='lastp',  
    p=50,                  
    leaf_rotation=90.,      
    leaf_font_size=8.,     
)
plt.axhline(y=max_d, c='k')
plt.show()
```

```python
x = pd.DataFrame(iris.data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
y = pd.DataFrame(iris.target, columns=['Target'])
```

```python
x.head()
```

```python
y.head()
```

```python
iris_k_mean_model = KMeans(n_clusters=3)
iris_k_mean_model.fit(x)
```

```python
print(iris_k_mean_model.labels_)
```

```python
print(iris_k_mean_model.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(14,6))

# Define colors array
colors = np.array(['red', 'green', 'blue'])

# Define predicted labels
predictedY = np.choose(iris_k_mean_model.labels_, [1, 0, 2]).astype(np.int64)

# Plot before classification
plt.subplot(1, 2, 1)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[y['Target']])
plt.title('Before classification')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10)], 
           labels=['Setosa', 'Versicolor', 'Virginica'])

# Plot model's classification
plt.subplot(1, 2, 2)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[predictedY])
plt.title("Model's classification")
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10)], 
           labels=['Setosa', 'Versicolor', 'Virginica'])

plt.show()
```

```python
sm.accuracy_score(predictedY, y['Target'])
```

```python
sm.confusion_matrix(predictedY, y['Target'])
```
